```python
import pandas as pd
```

```python
df = pd.read_csv('results.csv',
                sep=';',
                usecols=['race',
                         'cat',
                         'type',
                         'dist_km',
                         'gain_m',
                         'rank',
                         'total_entries',
                         'score',
                         'date'
                        ],
                parse_dates={'date_parsed':['date']},
                dayfirst=True,
                keep_date_col=False 
                )
```

```python
# New columns
df['year'] = df['date_parsed'].dt.year
df['month'] = df['date_parsed'].dt.month
df['day'] = df['date_parsed'].dt.day
# Set index
df.set_index('date_parsed', inplace=True)
```
